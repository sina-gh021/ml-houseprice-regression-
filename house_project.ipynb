```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
import types
%matplotlib inline
```

```python
 df = pd.read_csv("house_price.csv")
```

```python
df.describe()
```

```python
#cleaning nan  vlues
cdl = df [['Area','PriceUSD','Price','Address','Room']]
cdf=cdl.dropna().reset_index(drop=True)
```

```python
#cleaning data
k = 0 
for i in cdf['Area']:
    try:
        cdf['Area'][k]=float(i)
    except ValueError:
        cdf.drop([k], axis=0, inplace=True)
    k = k + 1

j = 0    
for i in cdf['Area']:
    if i == 0 :
        cdf.drop([j], axis=0, inplace=True)
    j=j+1

    



#j = 0 
#for i in cdf['Price']:
#    if type(i) != type(cdf['Price'][0]):
#        cdf.drop([j], axis=0, inplace=True)
 #   j=j+1
        
```

```python
#adding a new coloumn << gheymate metre morabae har khune >>
cdf['Price']=cdf['Price']/1000000000
cdf['M2pr']= cdf['Price'] / cdf['Area']

for i in cdf['M2pr'] : 
    if pd.isna(i):
        print(j)
        #cdf.drop([j], axis=0, inplace=True)
    j = j + 1

    
```

```python
        
areaprice={}
unadd = cdf['Address'].unique()
for i in unadd:
    condition = cdf['Address'] == i
    resault = cdf [condition].M2pr.values
    areaprice[i]=np.mean(resault)
areaprice=dict(sorted(areaprice.items(), key=lambda item: item[1]))    
def comp(row):
    j = 1
    for i in areaprice.keys():
        if row['Address'] == i :
            return j
        j = j+1
def com(row):
    
    for i in areaprice.keys():
        if row['Address'] == i :
            return areaprice.get(i)
#adding four new columns 
#pricep is the price that we predected basde on addressmean * Area
# diff = price - pricep << tafazol gheymat pishbini shode ba ghymat asli>>

cdf['Addressnumber']= cdf.apply(comp, axis=1)
cdf['Addressmean']= cdf.apply(com, axis=1)
cdf['Pricep']= cdf['Area'] * cdf['Addressmean']
cdf['diff']= cdf['Price'] - cdf['Pricep']
cdf.head(50)
```

```python

msk = np.random.rand(len(cdf)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
plt.scatter(cdf.Price, cdf.Pricep,  color='red')
plt.xlabel("pricep")
plt.ylabel("price")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
train_x = np.asanyarray(train[['Pricep']])
train_y = np.asanyarray(train[['Price']])

test_x = np.asanyarray(test[['Pricep']])
test_y = np.asanyarray(test[['Price']])

print (train_x[:3])
poly = PolynomialFeatures(degree=3)
train_x_poly = poly.fit_transform(train_x)
```

```python
clf = linear_model.LinearRegression()
train_y_ = clf.fit(train_x_poly, train_y)
# The coefficients
print ('Coefficients: ', clf.coef_)
print ('Intercept: ',clf.intercept_)
```

```python
plt.scatter(train.Pricep, train.Price,  color='blue')
XX = np.arange(0.0, 80, 0.1)
yy = clf.intercept_[0]+ clf.coef_[0][1]*XX+ clf.coef_[0][2]*np.power(XX, 2) +clf.coef_[0][3]*np.power(XX, 3)
plt.plot(XX, yy, 'red' )
plt.xlabel("Price")
plt.ylabel("pricep")
```

```python
from sklearn.metrics import r2_score

test_x_poly = poly.fit_transform(test_x)
test_y_ = clf.predict(test_x_poly)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y,test_y_ ) )
```
